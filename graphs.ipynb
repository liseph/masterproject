```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Twitter dataset

## Varying support values

```python
fig = plt.figure()
fig.suptitle('For sequence size = 25000', fontsize=12)
ax = plt.axes()
plt.grid(alpha=0.3, linestyle='dashed', lw=2, axis='y')
plt.xlabel('Support (in percentage)', fontsize=12)
plt.ylabel('Execution Time in ms', fontsize=12)
x = [0.11, 0.3, 0.5, 0.8, 1.0, 1.5, 10.0]
h = [3089.000, 2708.450, 2663.050, 2625.300, 2692.800, 2670.550, 2695.450]
st = [1818.300, 1099.000, 725.350, 664.550, 557.250, 382.150, 155.850]
line1, = plt.plot(x, h, marker='s', label='Hybrid')
line2, = plt.plot(x, st, marker='^', label='Spatial-Textual')
plt.legend(handles=[line1, line2], loc='best')
plt.savefig('graphs/twitterSupportVar.svg', format='svg', dpi=1200)
plt.show()
```

## Varying dataset size 

```python
fig = plt.figure()
fig.suptitle('For support = 0.8%', fontsize=12)
ax = plt.axes()
plt.grid(alpha=0.3, linestyle='dashed', lw=2, axis='y')
plt.xlabel('Number of sequences (K)', fontsize=12)
plt.ylabel('Execution Time in ms', fontsize=12)
x = [5, 15, 25, 50, 75, 100]
plt.xticks(x)
h = [521.750, 1601.850, 2611.900, 5237.000, 8785.500, 11689.400]
st = [115.750, 350.150, 616.350, 1175.400, 1988.750, 2413.850]
line1, = plt.plot(x, h, marker='s', label='Hybrid')
line2, = plt.plot(x, st, marker='^', label='Spatial-Textual')
plt.legend(handles=[line1, line2], loc='best')
plt.savefig('graphs/twitterSizeVar.svg', format='svg', dpi=1200)
plt.show()
```

## Varying location granularity

```python
fig = plt.figure()
fig.suptitle('For support = 0.8%, sequence size = 25000', fontsize=12)
ax = plt.axes()
plt.grid(alpha=0.3, linestyle='dashed', lw=2, axis='y')
plt.xlabel('Grid size (in unit size)', fontsize=12)
plt.ylabel('Execution Time in ms', fontsize=12)
x = [18.137790, 5.657243, 4.319923, 4.201380, 4.054036, 3.862433]
h = [2656.300, 2736.100, 2840.950, 2693.200, 2711.950, 2870.200]
st = [997.850, 650.450, 617.500, 673.600, 647.950, 700.850]
line1, = plt.plot(x, h, marker='s', label='Hybrid')
line2, = plt.plot(x, st, marker='^', label='Spatial-Textual')
plt.legend(handles=[line1, line2], loc='best')
plt.savefig('graphs/twitterLocVar.svg', format='svg', dpi=1200)
plt.show()
```
